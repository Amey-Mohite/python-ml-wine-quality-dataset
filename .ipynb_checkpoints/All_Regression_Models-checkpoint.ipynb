```python
data  = pd.read_csv('housingdata.csv')
```

```python
data.head()
```

```python
X =  data.drop(['MEDV'],1)
y = data['MEDV']
```

```python
X_train, X_test, y_train, y_test = tts(
    X,
    y,
    test_size=0.3,
    random_state=42)
```

```python
###################################### Modelss with function #################################
```

```python
def RMSLE(y, pred):
    return metrics.mean_squared_error(y, pred) ** 0.5
```

```python
def model(X_train,y_train,X_test,y_test,x):
    x.fit(X_train,y_train)
    print('Train set')
    pred=x.predict(X_train)
    print( ' Mean Squared error :{}'.format(mean_squared_error(y_train,pred)))
    print(' r2_score :{}'.format(r2_score(y_train,pred)))
    print('Tesst set')
    pred=x.predict(X_test)
    print(' Mean Squared error :{}'.format(mean_squared_error(y_test,pred)))
    print(' r2_score :{}'.format(r2_score(y_test,pred)))
    plt.scatter(y_test,pred,color='gray')
    plt.xlabel('Actual values')
    plt.ylabel('Predicted values')
    plt.plot(np.unique(y_test), np.poly1d(np.polyfit(y_test, pred, 1))(np.unique(y_test)))
    plt.text(0.6, 0.5, 'R-squared = %0.2f' % r2_score(y_test,pred))
    plt.show()
    df = pd.DataFrame({'Actual': y_test, 'Predicted': pred})
    df1 = df.head(30)
    df1.plot(kind='bar',figsize=(10,8))
    plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
    plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
    plt.show()
    visualizer = ResidualsPlot(x)
    visualizer.fit(X_train, y_train)
    visualizer.score(X_test, y_test)
    visualizer.poof()
```

```python
def crossvalidation(model,method):
    scoring = 'r2'
    results = cross_val_score(model, X, y, cv=method)
    result = results.mean()
    print("model : ",model," method : ",method,"--",result)
```

```python
#normal Model
lr = LinearRegression()
model(X_train,y_train,X_test,y_test,lr)

#Cross Validation Model 
kfold = KFold(n_splits=10, random_state=42)
crossvalidation(lr,kfold)
rkf = RepeatedKFold(n_splits=2, n_repeats=2, random_state=42)
crossvalidation(lr,rkf)
loo = LeaveOneOut()
crossvalidation(lr,loo)
ss = ShuffleSplit(n_splits=5, test_size=0.25, random_state=0)
crossvalidation(lr,ss)
# skf = StratifiedKFold(n_splits=3,)
# crossvalidation(lr,skf)
selector = RFECV(lr, cv=5)
selector = selector.fit(X, y)
print("RFECV : ",selector.score(X,y))
```

```python
#normal model

dtm = DecisionTreeRegressor(max_depth=4, min_samples_split=5,max_leaf_nodes=10)
model(X_train,y_train,X_test,y_test,dtm)

#Cross Validation Score 
kfold = KFold(n_splits=10, random_state=42)
crossvalidation(dtm,kfold)
rkf = RepeatedKFold(n_splits=2, n_repeats=2, random_state=42)
crossvalidation(dtm,rkf)
loo = LeaveOneOut()
crossvalidation(dtm,loo)
ss = ShuffleSplit(n_splits=5, test_size=0.25, random_state=0)
crossvalidation(dtm,ss)
# skf = StratifiedKFold(n_splits=3,)
# crossvalidation(lr,skf)
selector = RFECV(dtm, cv=5)
selector = selector.fit(X, y)
print("RFECV : ",selector.score(X,y))

#Grid Search CV or Randomise Search CV

param_grid = {"criterion": ["mse", "mae"],
              "min_samples_split": [10, 20, 40],
              "max_depth": [2, 6, 8],
              "min_samples_leaf": [20, 40, 100],
              "max_leaf_nodes": [5, 20, 100],
              }
grid_cv_dtm = GridSearchCV(dtm, param_grid, cv=5)
model(X_train,y_train,X_test,y_test,grid_cv_dtm)
```
